**Q2** IF/ELIF/ELSE statements

<font size = "3">

- In the beginning of the code cell below, two strings are created, where the first word is one of three random choices: 
```python
    title_py = "Introduction/Intro/Guide to Python"
    title_r = "Introduction/Intro/Guide to R"
```

- Assign the string "Introduction" to a variable named `keyword`
- Create an if/elif/else block that tests whether `keyword` appears in the two strings.  

- Depending on the outcome (appears in both strings, appears in `title_py` only, appears in `title_r` only, doesn't appear in either string) print an appropriate message to the screen.

HINT: If the creation of `title_py` and `title_r` is confusing, add print statements to view the two strings and run the cell 3-4 times to see some of the possible situations

````python
# ### Code to create strings. DO NOT CHANGE!!

import random  # load random library to choose words randomly

first_word_choices = ["Introduction", "Intro", "Guide"]  
# list of possible first words for the titles

selected_words = random.choices(first_word_choices, k = 2)  
# randomly pick 2 words from the list

title_py = selected_words[0] + " to Python"  
# create a string like "Intro to Python"

title_r = selected_words[1] + " to R"  
# create a string like "Guide to R"


### Your solution below

keyword = "Introduction"  
# the word we want to search for inside the titles


if keyword in title_py:  
    # check if "Introduction" is inside the Python title
    
    if keyword in title_r:  
        # check if it is ALSO inside the R title
        print('Keyword appears in both strings')
        
    else:  
        # means it is in Python title but NOT in R title
        print('Keyword appears in title_py')

elif keyword in title_r:  
    # this runs if keyword is NOT in Python title
    # but IS in R title
    print('Keyword appears in title_r')

else:  
    # keyword is not in either title
    print('Keyword is not in either string')
````

**Q5** Generate a vector of values

<font size = "3">

- Below, two lists are defined containing values of $x_k$ and $y_k$

```python
    coord_x = [1,4, 0, -1, 20, 40]
    coord_y = [2,2, 0, 10, 0, 1]
```

- Create a vector (**vector always means numpy array**) ```hypotenuse_sqr``` with elements $h_k$ defined by the formula

$$ h_k = x_k^2 + y_k^2$$

- You **should not** define it manually. So the following is **not** an acceptable answer:
```python
    hypoteneuse_sqr = np.array([5, 20, 0, 101, 400, 1601])
```

Hint: Most mathematical operations are defined for NumPy arrays, which allows you to avoid loops

````python
coord_x = [1, 4, 0, -1, 20, 40]  
# list of x-coordinates

coord_y = [2, 2, 0, 10, 0, 1]  
# list of y-coordinates


### Multiple possible solutions given below

#### Solution 1 (element-wise operations with NumPy)
hyp_sqr_1 = np.array(coord_x)**2 + np.array(coord_y)**2  
# convert both lists to NumPy arrays
# square each value element-by-element
# add x^2 + y^2 for each coordinate pair
# result: array of squared hypotenuse values


#### Solution 2 (element-wise operations with NumPy)
x_array = np.array(coord_x)  
# convert x list to NumPy array

y_array = np.array(coord_y)  
# convert y list to NumPy array

hyp_sqr_2 = x_array**2 + y_array**2  
# compute x^2 + y^2 element-wise (same result as Solution 1)


#### Solution 3 (Using loop and ".append")
N = len(coord_x)  
# number of coordinate points

hyp_list = []  
# start with an empty list to store results

for k in range(N):  
    # loop through each index position
    hyp_list.append(coord_x[k]**2 + coord_y[k]**2)  
    # compute x^2 + y^2 for that point and add to list

hyp_sqr_3 = np.array(hyp_list)  
# convert the final list into a NumPy array


#### Solution 4 (Using loop and pre-made zero array)
N = len(coord_x)  
# number of coordinate points

hyp_sqr_4 = np.zeros(N)  
# create a NumPy array of zeros with length N
# this will be filled with correct values in the loop

for loop_var in range(N):  
    # go through each index
    hyp_sqr_4[loop_var] = coord_x[loop_var]**2 + coord_y[loop_var]**2  
    # replace each zero with x^2 + y^2 value


# check answer
print(hyp_sqr_1)  
print(hyp_sqr_2)  
print(hyp_sqr_3)  
print(hyp_sqr_4)  
# all four printed arrays should be the SAME
````

**Q6** Sum all positive elements

<font size = "3">

Below, we create a random `numpy` array with positive and negative numbers. Using a `for` loop with a `continue` statement, compute the sum of all the **positive** elements of the array

````python
N = 25  
# number of random values to generate

random_array = np.random.uniform(low = -5, high = 10, size = N)  
# create a NumPy array of 25 random numbers
# values range from -5 to 10


###### Solution 1
pos_sum_1 = 0  
# start a variable to store the sum of positive numbers

for val in random_array:  
    # loop directly through each value in the array
    
    if val <= 0:  
        # if the value is zero or negative
        continue  
        # skip this value and move to the next loop iteration
    
    pos_sum_1 = pos_sum_1 + val  
    # add the positive value to the running sum


###### Solution 2
pos_sum_2 = 0  
# start another sum variable (same purpose as above)

for i in range(N):  
    # loop through index positions 0  24
    
    if random_array[i] <= 0:  
        # check if the value at this index is zero or negative
        continue  
        # skip if not positive
    
    pos_sum_2 = pos_sum_2 + random_array[i]  
    # add the positive value to the sum


print(pos_sum_1)  
print(pos_sum_2)  
# both printed sums should be the SAME
````
